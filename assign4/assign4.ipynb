```python
import pandas as pd

df = pd.read_csv("scores.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
print(df.columns)
```

```python
correlation = df['Hours'].corr(df['Scores'])
print("Correlation:", correlation)
```

```python
# Clean column names
df.columns = df.columns.str.strip()

# Rename columns
df.columns = ['study_hours', 'exam_score']
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['study_hours'], df['exam_score'])
plt.title("Study Hours vs Exam Score")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.show()
```

```python
df['exam_score'].hist()
plt.title("Distribution of Exam Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

```python
from scipy.stats import pearsonr

corr, p_value = pearsonr(df['study_hours'], df['exam_score'])

print("Correlation:", corr)
print("P-value:", p_value)
```

```python
if p_value < 0.05:
    print("Reject Null Hypothesis → Study time affects scores")
else:
    print("Fail to Reject Null → No significant effect")
```
