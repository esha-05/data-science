```python
import pandas as pd

df = pd.read_csv("Simple 100 Student Marks.csv")
```

```python
df.head()
```

```python
# Check structure
df.info()

# Summary stats
df.describe()
```

```python
# Select only numeric columns
marks_df = df.select_dtypes(include='number')

# Mean
mean = marks_df.mean()

# Median
median = marks_df.median()

# Mode
mode = marks_df.mode()

# Standard Deviation
std_dev = marks_df.std()

print("Mean:\n", mean)
print("\nMedian:\n", median)
print("\nMode:\n", mode)
print("\nStandard Deviation:\n", std_dev)
```

```python
import matplotlib.pyplot as plt

marks_df.hist(figsize=(12,8))
plt.suptitle("Distribution of Student Marks")
plt.show()
```

```python
marks_df.plot(kind='box', figsize=(10,6))
plt.title("Boxplot of Student Marks")
plt.show()
```

```python
mean.plot(kind='bar', figsize=(8,5))
plt.title("Average Marks per Subject")
plt.ylabel("Marks")
plt.show()
```

```python
df['total'] = marks_df.sum(axis=1)
```

```python
topper = df.loc[df['total'].idxmax()]
print(topper)
```

```python
lowest = df.loc[df['total'].idxmin()]
print(lowest)
```

## 📊 Interpretation of Results

### 1. Overall Performance
The average marks indicate that students are performing moderately well across subjects. Some subjects have higher average scores, suggesting they may be easier or better understood.

### 2. Spread of Marks
The standard deviation shows how spread out the marks are. Subjects with higher deviation indicate varied student performance, while lower deviation suggests consistency.

### 3. Distribution of Scores
From the histograms, most students' scores fall within a central range, indicating a fairly normal distribution. There are fewer extreme low or high scores.

### 4. Outliers
The boxplot reveals a few outliers, which represent students with exceptionally high or low marks.

### 5. Subject Comparison
Some subjects have higher average scores compared to others, indicating relative ease or stronger student performance in those areas.

### 6. Top and Low Performers
Using total marks, we identified the highest and lowest scoring students, helping highlight performance extremes.
